```python
# Aggregating the data by taking the average for each event

def aggregate_data(data, status):
    """ Aggregate the data by averaging across individuals for each event. """
    # Find the maximum number of events for any individual
    max_events = max(len(data[ind][status]) for ind in data)

    # Initialize a list to store the sum and count for each event
    event_sums = [0] * max_events
    event_counts = [0] * max_events

    # Sum up the rows and count the number of individuals for each event
    for ind in data:
        for event, rows in enumerate(data[ind][status]):
            event_sums[event] += rows
            event_counts[event] += 1

    # Compute the average for each event
    averages = [event_sums[i] / event_counts[i] if event_counts[i] != 0 else 0 for i in range(max_events)]
    return averages

# Aggregate the data for 'not obeyed' and 'obeyed'
averages_not_obeyed = aggregate_data(data, 'not obeyed')
averages_obeyed = aggregate_data(data, 'obeyed')

# Now, let's plot the aggregated data

# Prepare the subplots with larger size
fig, axes = plt.subplots(1, 2, figsize=(15, 8))
fig.suptitle('Aggregated Rows Count for Each Event')

# Plot for 'not obeyed'
axes[0].plot(averages_not_obeyed, label='Average Not Obeyed', color='black')
axes[0].set_title('Cumulative Average Days Until Next Robbey After Unsuccesful Robbery')
axes[0].set_xlabel('Occurences')
axes[0].set_ylabel('Average Days Until Next Robbery')
axes[0].legend()

# Plot for 'obeyed'
axes[1].plot(averages_obeyed, label='Average Obeyed', color='black')
axes[1].set_title('Cumulative Average Days Until Next Robbey After Succesful Robbery')
axes[1].set_xlabel('Occurences')
axes[1].set_ylabel('Average Days Until Next Robbery')
axes[1].legend()

# Show the plots
plt.show()
```
